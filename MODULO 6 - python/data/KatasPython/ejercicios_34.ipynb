```python
"""
Ejercicio 34. Crea la clase Arbol , define un árbol genérico con un tronco y ramas como atributos. Los métodos disponibles son:
crecer_tronco , nueva_rama , crecer_ramas , quitar_rama e info_arbol . El objetivo es implementar estos métodos para
manipular la estructura del árbol.
"""
class Arbol:
    def __init__(self):
        """
        Inicializa un árbol con tronco de longitud 1 y lista vacía de ramas.
        """
        self.tronco = 1
        self.ramas = []

    def crecer_tronco(self):
        """
        Aumenta la longitud del tronco en 1 unidad.
        """
        self.tronco += 1
        print(f"Tronco creció. Longitud actual: {self.tronco}")

    def nueva_rama(self):
        """
        Agrega una nueva rama de longitud 1 a la lista de ramas.
        """
        self.ramas.append(1)
        print(f"Nueva rama añadida. Número total de ramas: {len(self.ramas)}")

    def crecer_ramas(self):
        """
        Aumenta en 1 la longitud de todas las ramas existentes.
        """
        self.ramas = [rama + 1 for rama in self.ramas]
        print(f"Todas las ramas crecieron. Longitudes actuales: {self.ramas}")

    def quitar_rama(self, indice):
        """
        Elimina una rama en la posición especificada.
        """
        if 0 <= indice < len(self.ramas):
            rama_eliminada = self.ramas.pop(indice)
            print(f"Rama en posición {indice} eliminada (longitud {rama_eliminada}).")
        else:
            print(f"Error: índice {indice} fuera de rango. No se eliminó ninguna rama.")

    def info_arbol(self):
        """
        Devuelve información sobre la longitud del tronco, número de ramas y longitudes de las mismas.
        """
        info = (
            f"Longitud del tronco: {self.tronco}\n"
            f"Número de ramas: {len(self.ramas)}\n"
            f"Longitudes de las ramas: {self.ramas}"
        )
        return info

# Paso 1: Crear un árbol
mi_arbol = Arbol()

# Paso 2: Hacer crecer el tronco del árbol una unidad
mi_arbol.crecer_tronco()

# Paso 3: Añadir una nueva rama al árbol
mi_arbol.nueva_rama()

# Paso 4: Hacer crecer todas las ramas del árbol una unidad
mi_arbol.crecer_ramas()

# Paso 5: Añadir dos nuevas ramas al árbol
mi_arbol.nueva_rama()
mi_arbol.nueva_rama()

# Paso 6: Retirar la rama situada en la posición 2
mi_arbol.quitar_rama(2)

# Paso 7: Obtener información sobre el árbol
print(mi_arbol.info_arbol())
```
