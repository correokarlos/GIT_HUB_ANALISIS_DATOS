```python
"""
Ejercicio 11. Escribe un programa que pida al usuario que introduzca su edad. Si el usuario ingresa un valor no numérico o un
valor fuera del rango esperado (por ejemplo, menor que 0 o mayor que 120), maneja las excepciones adecuadamente.
"""
class EdadFueraDeRango(Exception):
    """Excepción personalizada para valores inválidos."""
    pass

def solicitar_edad():
    """_summary_

    Raises:
        EdadFueraDeRango: _description_
    """
    edad = int(input("Por favor, introduce tu edad: "))
    if edad < 0 or edad > 120:
            raise EdadFueraDeRango("La edad debe estar entre 0 y 120.")   
        
    print(f"Tienes {edad} años.")

try:
    solicitar_edad()
except ValueError as e:
    print(f"Error: {e}")
except EdadFueraDeRango as e:
    print(f"Error: {e}")
```

```python
"""
Ejercicio 12: Genera una función que al recibir una frase devuelva una lista con la longitud de cada palabra. Usa la función map()
"""

def longitudes_palabras(frase):
    """Esta función recibe una frase y devuelve una lista con la longitud de cada palabra.

    Args:
        frase (str): Frase introducida por el usuario.

    Returns:
        list: Lista con las longitudes de cada palabra.
    """
    palabras = frase.split()
    longitudes = list(map(len, palabras))
    return longitudes   

frase_usuario = input("Introduce una frase: ")
longitudes = longitudes_palabras(frase_usuario) 
print("Longitudes de las palabras:", longitudes)
```

```python
"""
Ejercicio 13. Genera una función la cual, para un conjunto de caracteres, devuelva una lista de tuplas con cada letra en
mayúsculas y minúsculas. Las letras no pueden estar repetidas .Usa la función map()
"""
def letras_mayus_minus(conjunto):
    """
    Esta función recibe un conjunto de caracteres y devuelve una lista de tuplas con cada letra en mayúsculas y minúsculas.

    Args:
        conjunto (str): Conjunto de caracteres introducido por el usuario.

    Returns:
        list: Lista de tuplas con letras en mayúsculas y minúsculas.
    """
    caracteres_unicos = set(conjunto)
    
    
    def mayus_minus(c):
        return (c.upper(), c.lower())
    
    resultado = list(map(mayus_minus, caracteres_unicos))
    return resultado

conjunto_usuario = input("Introduce un conjunto de caracteres: ")
tuplas_letras = letras_mayus_minus(conjunto_usuario)
print("Lista de tuplas con letras en mayúsculas y minúsculas:", tuplas_letras)
```

```python
"""
Ejercicio 14. Crea una función que retorne las palabras de una lista de palabras que comience con una letra en especifico. Usa la
función filter()
"""
def filtrar_palabras_por_letra(palabras, letra):
    """Esta función filtra las palabras que comienzan con una letra específica.

    Args:
        palabras (list): Lista de palabras introducida por el usuario.
        letra (str): Letra específica para filtrar las palabras.

    Returns:
        list: Lista de palabras que comienzan con la letra especificada.
    """
    
    def comienza_con_letra(palabra):
        return palabra.lower().startswith(letra.lower())
    
    palabras_filtradas = list(filter(comienza_con_letra, palabras))
    return palabras_filtradas

lista_palabras = input("Introduce una lista de palabras separadas por espacios: ").split()
letra_especifica = input("Introduce la letra para filtrar las palabras: ")
palabras_filtradas = filtrar_palabras_por_letra(lista_palabras, letra_especifica)
print("Palabras que comienzan con la letra especificada:", palabras_filtradas)
```

```python
"""
Ejercicio 15. Crea una función lambda que sume 3 a cada número de una lista dada.
"""
def sumar_tres_a_lista(numeros):
    """Esta función suma 3 a cada número de una lista dada usando una función lambda.

    Args:
        numeros (list): Lista de números introducida por el usuario.

    Returns:
        list: Nueva lista con cada número incrementado en 3.
    """
    sumar_tres = lambda x: x + 3
    nueva_lista = list(map(sumar_tres, numeros))
    return nueva_lista

try:
    lista_numeros = list(map(int, input("Introduce una lista de números separados por espacios: ").split()))
    nueva_lista_numeros = sumar_tres_a_lista(lista_numeros)
    print("Nueva lista con cada número incrementado en 3:", nueva_lista_numeros)
except ValueError as e:
    print(f"Error de valor: {e}")
except Exception as e:
    print(f"Ocurrió un error inesperado: {e}")
```

```python
"""
Ejercicio 16. Escribe una función que tome una cadena de texto y un número entero n como parámetros y devuelva una lista de
todas las palabras que sean más largas que n. Usa la función filter()
"""

def palabras_mas_largas_que_n(frase, n):
    """Esta función devuelve una lista de palabras que son más largas que n.

    Args:
        frase (str): Cadena de texto introducida por el usuario.
        n (int): Número entero para comparar la longitud de las palabras.

    Returns:
        list: Lista de palabras más largas que n.
    """
    palabras = frase.split()
    
    def es_mas_larga(palabra):
        return len(palabra) > n
    
    palabras_filtradas = list(filter(es_mas_larga, palabras))
    return palabras_filtradas

frase_usuario = input("Introduce una frase: ")
try:
    n = int(input("Introduce un número entero n: "))
    palabras_filtradas = palabras_mas_largas_que_n(frase_usuario, n)
    print("Palabras más largas que n:", palabras_filtradas)
except ValueError as e:
    print(f"Error de valor: {e}")
except Exception as e:
    print(f"Ocurrió un error inesperado: {e}")
```

```python
"""
Ejercicio 17:Crea una función que tome una lista de dígitos y devuelva el número correspondiente. Por ejemplo, [5,7,2]
corresponde al número quinientos setenta y dos (572). Usa la función reduce()
"""
def lista_a_numero(digitos):
    """Esta función convierte una lista de dígitos en el número correspondiente.

    Args:
        digitos (list): Lista de dígitos introducida por el usuario.

    Returns:
        int: Número correspondiente a la lista de dígitos.
    """
    from functools import reduce
    
    numero = reduce(lambda x, y: x * 10 + y, digitos)
    return numero

try:
    lista_digitos = [4, 6, 7]
    numero_resultante = lista_a_numero(lista_digitos)
    print("El número correspondiente es:", numero_resultante)   
except ValueError as e:
    print(f"Error de valor: {e}")
except Exception as e:
    print(f"Ocurrió un error inesperado: {e}")
```

```python
"""
Ejercicio 18. Escribe un programa en Python que cree una lista de diccionarios que contenga información de estudiantes
(nombre, edad, calificación) y use la función filter para extraer a los estudiantes con una calificación mayor o igual a
90. Usa la función filter()
"""
def estudiantes_con_alta_calificacion(estudiantes):
    """Esta función filtra a los estudiantes con una calificación mayor o igual a 90.

    Args:
        estudiantes (list): Lista de diccionarios con información de estudiantes.

    Returns:
        list: Lista de estudiantes con calificación mayor o igual a 90.
    """
    
    def tiene_alta_calificacion(estudiante):
        return estudiante['calificación'] >= 90
    
    estudiantes_filtrados = list(filter(tiene_alta_calificacion, estudiantes))
    return estudiantes_filtrados
lista_estudiantes = [
    {'nombre': 'Ana', 'edad': 20, 'calificación': 95},
    {'nombre': 'Luis', 'edad': 22, 'calificación': 85},
    {'nombre': 'Marta', 'edad': 19, 'calificación': 92},
    {'nombre': 'Carlos', 'edad': 21, 'calificación': 88}]
estudiantes_destacados = estudiantes_con_alta_calificacion(lista_estudiantes)
print("Estudiantes con calificación mayor o igual a 90:", estudiantes_destacados)
```

```python
"""
Ejercicio 19. Crea una función lambda que filtre los números impares de una lista dada.
"""
def filtrar_numeros_impares(numeros):
    """Esta función filtra los números impares de una lista dada usando una función lambda.

    Args:
        numeros (list): Lista de números introducida por el usuario.

    Returns:
        list: Lista de números impares.
    """
    es_impar = lambda x: x % 2 != 0
    numeros_impares = list(filter(es_impar, numeros))
    return numeros_impares

try:
    lista_numeros = list(map(int, input("Introduce una lista de números separados por espacios: ").split()))
    numeros_impares = filtrar_numeros_impares(lista_numeros)
    print("Números impares en la lista:", numeros_impares)
except ValueError as e:
    print(f"Error de valor: {e}")
except Exception as e:
    print(f"Ocurrió un error inesperado: {e}")
```

```python
"""
Ejercicio 20. Para una lista con elementos tipo integer y string obtén una nueva lista sólo con los valores int. Usa la función
filter()
"""
def filtrar_enteros(elementos):
    """Esta función filtra los valores enteros de una lista que contiene elementos de tipo integer y string.

    Args:
        elementos (list): Lista con elementos de tipo integer y string.

    Returns:
        list: Nueva lista sólo con los valores int.
    """
    es_entero = lambda x: isinstance(x, int)
    enteros_filtrados = list(filter(es_entero, elementos))
    return enteros_filtrados

lista_mixta = [10, 'hola', 25, 'mundo', 42, 'python', 7]
enteros = filtrar_enteros(lista_mixta)
print("Lista sólo con valores int:", enteros)
```
