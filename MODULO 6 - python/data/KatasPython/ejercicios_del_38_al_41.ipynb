```python
"""
Ejercicio 38. Genera un programa que nos diga si es de noche, de día o tarde según la hora proporcionada por el usuario.
"""
def determinar_momento_del_dia(hora):
    """
    Determina si es día, tarde o noche según la hora.

    Args:
        hora (int): Hora en formato 24 horas (0-23).

    Returns:
        str: "Mañana/Día", "Tarde" o "Noche"
    """
    if 6 <= hora < 12:
        return "Mañana/Día"
    elif 12 <= hora < 18:
        return "Tarde"
    else:
        return "Noche"

# Solicitar hora al usuario con control de errores
try:
    hora_usuario = int(input("Introduce la hora actual (0-23): "))
    if hora_usuario < 0 or hora_usuario > 23:
        raise ValueError("La hora debe estar entre 0 y 23.")

    momento = determinar_momento_del_dia(hora_usuario)
    print(f"Según la hora {hora_usuario}, es: {momento}")

except ValueError as e:
    print(f"Error: {e}")
```

```python
"""
Ejercicio 39. Escribe un programa que determine qué calificación en texto tiene un alumno en base a su calificación numérica.
"""
def calificacion_texto(nota):
    """
    Convierte una calificación numérica en una calificación en texto.

    Args:
        nota (float): Calificación numérica (0-10).

    Returns:
        str: Calificación en texto.
    """
    if nota < 0 or nota > 10:
        return "Nota inválida"
    elif nota < 5:
        return "Insuficiente"
    elif nota < 6:
        return "Suficiente"
    elif nota < 7:
        return "Bien"
    elif nota < 9:
        return "Notable"
    else:
        return "Sobresaliente"

# Solicitar calificación al usuario con control de errores
try:
    nota_usuario = float(input("Introduce la calificación del alumno (0-10): "))
    texto = calificacion_texto(nota_usuario)
    print(f"La calificación en texto es: {texto}")
except ValueError:
    print("Error: Debes introducir un número válido.")
```

```python
"""
Ejercicio 40. Escribe una función que tome dos parámetros: figura (una cadena que puede ser "rectangulo" , "circulo" o
"triangulo" ) y datos (una tupla con los datos necesarios para calcular el área de la figura).
"""
import math

def calcular_area(figura, datos):
    """
    Calcula el área de una figura geométrica.

    Args:
        figura (str): Tipo de figura ("rectangulo", "circulo", "triangulo").
        datos (tuple): Datos necesarios para calcular el área.
                       - Rectángulo: (base, altura)
                       - Círculo: (radio,)
                       - Triángulo: (base, altura)

    Returns:
        float: Área de la figura.
    """
    figura = figura.lower()

    if figura == "rectangulo":
        if len(datos) != 2:
            raise ValueError("Rectángulo requiere 2 datos: base y altura.")
        base, altura = datos
        return base * altura

    elif figura == "circulo":
        if len(datos) != 1:
            raise ValueError("Círculo requiere 1 dato: radio.")
        radio = datos[0]
        return math.pi * radio ** 2

    elif figura == "triangulo":
        if len(datos) != 2:
            raise ValueError("Triángulo requiere 2 datos: base y altura.")
        base, altura = datos
        return (base * altura) / 2

    else:
        raise ValueError("Figura no válida. Debe ser 'rectangulo', 'circulo' o 'triangulo'.")

# Rectángulo: base=5, altura=3
area_rectangulo = calcular_area("rectangulo", (5, 3))
print(f"Área del rectángulo: {area_rectangulo}")

# Círculo: radio=4
area_circulo = calcular_area("circulo", (4,))
print(f"Área del círculo: {area_circulo:.2f}")

# Triángulo: base=6, altura=2
area_triangulo = calcular_area("triangulo", (6, 2))
print(f"Área del triángulo: {area_triangulo}")
```

```python
"""
Ejercicio 41. En este ejercicio, se te pedirá que escribas un programa en Python que utilice condicionales para determinar el
monto final de una compra en una tienda en línea, después de aplicar un descuento.
"""
# Solicitar el precio original
try:
    precio_original = float(input("Introduce el precio original del artículo: "))
    if precio_original < 0:
        raise ValueError("El precio no puede ser negativo.")

    # Preguntar si tiene cupón
    respuesta_cupon = input("¿Tienes un cupón de descuento? (sí/no): ").strip().lower()

    if respuesta_cupon == "sí" or respuesta_cupon == "si":
        # Solicitar valor del cupón
        descuento = float(input("Introduce el valor del cupón de descuento: "))
        if descuento <= 0:
            print("El cupón no es válido. No se aplicará ningún descuento.")
            precio_final = precio_original
        elif descuento > precio_original:
            print("El descuento no puede superar el precio. Se aplicará descuento máximo igual al precio.")
            precio_final = 0
        else:
            precio_final = precio_original - descuento
    elif respuesta_cupon == "no":
        precio_final = precio_original
    else:
        print("Respuesta no válida. Se considerará que no hay cupón.")
        precio_final = precio_original

    print(f"El precio final de la compra es: {precio_final:.2f}€")

except ValueError as e:
    print(f"Error: {e}")
```
