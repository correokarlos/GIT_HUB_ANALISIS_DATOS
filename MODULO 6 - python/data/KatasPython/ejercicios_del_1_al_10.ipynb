```python
""" 
Ejerccio 1: Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias
de cada letra en la cadena. Los espacios no deben ser considerados.
"""
def contar_frecuencias(cadena):
    """
    Contar frecuencias de letras en una cadena de texto.

    Args:
        cadena (str): cadena de texto a iterar

    Returns:
        dicc: estructura de datos con las letras y sus frecuencias
    """
    frecuencias = {}
    for letra in cadena:
        if letra != ' ':
            if letra in frecuencias:
                frecuencias[letra] += 1
            else:
                frecuencias[letra] = 1
    return frecuencias

# Ejemplo de uso
texto = "hola mundo"    
resultado = contar_frecuencias(texto)
print(resultado)  # Salida esperada: {'h': 1, 'o': 2, 'l': 1, 'a': 1, 'm': 1, 'u': 1, 'n': 1, 'd': 1}
```

```python
"""
Ejercicio 2. Dada una lista de números, obtén una nueva lista con el doble de cada valor. Usa la función map()
"""
def doblar_numeros(numero):
    """
    Esta función recibe una lista de números y devuelve una nueva lista con el doble de cada valor.

    Args:
        numero (int): numero original a doblar

    Returns:
        int: nuemro doblado
    """
    return numero * 2


# definimos una lista de números sobre la que queramos aplicar dicha función.
numeros = [1,2,3,4,5]
print(f"la lista original antes de elevar cada elemento al cuadrado es: {numeros}")

# utilizamos el map para ejecutar dicha función a todos los elementos de nuestra lista numbers.
numeros_doblados = map(doblar_numeros,numeros)
print(f"El resultado de la función map es: {list(numeros_doblados)}")

```

```python
"""
5. Ecribe una función que tome una lista de números como parámetro y un valor opcional nota_aprobado, que por
defecto es 5. La función debe calcular la media de los números en la lista y determinar si la media es mayor o igual
que nota aprobado. Si es así, el estado será "aprobado", de lo contrario, será "suspenso". La función debe devolver
una tupla que contenga la media y el estado.
"""

def evaluar_nota(numeros, nota_aprobado=5):
    """
    Esta función toma una lista de números como parámetro y un valor opcional nota_aprobado, que por
    defecto es 5. La función calcula la media de los números en la lista y determina si la media es mayor o igual
    que nota aprobado. Si es así, el estado será "aprobado", de lo contrario, será "suspenso". La función devuelve
    una tupla que contiene la media y el estado.

    Args:
        numeros (list): lista de números
        nota_aprobado (int, optional): nota de aprobado. Defaults to 5.

    Returns:
        tuple: tupla con la media y el estado
    """
    media = sum(numeros) / len(numeros)
    estado = "aprobado" if media >= nota_aprobado else "suspenso"
    return (media, estado)

# Ejemplo de uso
notas = [6,7,8,4,5]
resultado = evaluar_nota(notas) 
print(resultado)  # Salida esperada: (6.0, 'aprobado')

```

```python
"""
Ejercicio 6: Escribe una función que calcule el factorial de un número de manera recursiva.
"""
def factorial(n):
    """
    Esta función calcula el factorial de un número de manera recursiva.

    Args:
        n (int): número para calcular el factorial

    Returns:
        int: factorial del número
    """
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
    
# Ejemplo de uso
numero = 5
resultado = factorial(numero)
print(f"El factorial de {numero} es {resultado}")  # Salida esperada: El factorial de 5 es 120
```

```python
"""
Ejercicio 7: Genera una función que convierta una lista de tuplas a una lista de strings. Usa la función map()
"""
def tuplas_a_strings(tupla):
    """
    Esta función convierte una tupla a un string.

    Args:
        tupla (tuple): tupla a convertir

    Returns:
        str: string resultante de la tupla
    """
    list_strings = []

    for elemento in tupla:
        list_strings.append(str(elemento))  

    return ' '.join(list_strings)

# definimos una lista de tuplas sobre la que queramos aplicar dicha función.
tuplas = [(1, 'hola'), (2, 'mundo'), (3, '!')]
print(f"la lista original de tuplas es: {tuplas}")
# utilizamos el map para ejecutar dicha función a todos los elementos de nuestra lista tuplas.
strings = map(tuplas_a_strings, tuplas) 
print(f"El resultado de la función map es: {list(strings)}")
```

```python
"""
Ejercicio 9: Escribe una función que tome una lista de nombres de mascotas como parámetro y devuelva una nueva lista
excluyendo ciertas mascotas prohibidas en España. La lista de mascotas a excluir es ["Mapache", "Tigre",
"Serpiente Pitón", "Cocodrilo", "Oso"].Usa la función filter()
"""

def filtrar_mascotas(mascotas):
    """
    Esta función toma una lista de nombres de mascotas como parámetro y devuelve una nueva lista
    excluyendo ciertas mascotas prohibidas en España.

    Args:
        mascotas (list): lista de nombres de mascotas

    Returns:
        list: lista de mascotas permitidas
    """
    mascotas_prohibidas = ["Mapache", "Tigre", "Serpiente Pitón", "Cocodrilo", "Oso"]
    mascotas_permitidas = []

    for mascota in mascotas:
        if mascota not in mascotas_prohibidas:
            mascotas_permitidas.append(mascota)

    return mascotas_permitidas

# Ejemplo de uso
lista_mascotas = ["Perro", "Gato", "Mapache", "Loro", "Tigre", "Hamster"]
resultado = filtrar_mascotas(lista_mascotas)    
print(resultado)  # Salida esperada: ['Perro', 'Gato', 'Loro', 'Hamster']


```
