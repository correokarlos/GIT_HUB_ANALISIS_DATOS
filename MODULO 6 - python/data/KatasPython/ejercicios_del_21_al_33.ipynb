```python
from functools import reduce
"""
Ejercicio 21. Crea una función que calcule el cubo de un número dado mediante una función lambda
"""
def calcular_cubo(numero):
    cubo = lambda x: x ** 3
    return cubo(numero)     

# Ejemplo de uso
numero = 4
resultado = calcular_cubo(numero)
print(f"El cubo de {numero} es {resultado}")
```

```python
"""
Ejercicio 22. 22. Dada una lista numérica, obtén el producto total de los valores de dicha lista.Usa la función reduce() .
"""

def multiplicar(x, y):
    """Devuelve el producto de dos números."""
    return x * y

def producto_total(lista_numeros):
    """
    Calcula el producto total de todos los elementos de una lista numérica.

    Args:
        lista_numeros (list): Lista de números.

    Returns:
        int o float: Producto total de los valores de la lista.
    """
    return reduce(multiplicar, lista_numeros)

# Ejemplo de uso

numeros = [22, 8, 9]  
resultado = producto_total(numeros)
print(f"El producto total de los valores es: {resultado}")
```

```python
"""
Ejercicio 23. Concatena una lista de palabras.Usa la función reduce() .
"""

def concatenar(palabra1, palabra2):
    """Une dos palabras en una sola cadena."""
    return palabra1 + palabra2

def concatenar_lista(palabras):
    """
    Concatena todas las palabras de una lista en una sola cadena.

    Args:
        palabras (list): Lista de palabras (strings).

    Returns:
        str: Cadena resultante de concatenar todas las palabras.
    """
    return reduce(concatenar, palabras)

# Ejemplo de uso

lista_palabras = ["Hola", "como", "va", "el", "dia"]
resultado = concatenar_lista(lista_palabras)
print("Resultado de la concatenación:", resultado)

```

```python
"""
Ejercicio 24. Calcula la diferencia total en los valores de una lista. Usa la función reduce() .
"""
def diferencia_total(lista_numeros):
    """
    Calcula la diferencia total de los valores de una lista usando reduce().

    Args:
        lista_numeros (list): Lista de números.

    Returns:
        int o float: Resultado de restar todos los valores de izquierda a derecha.
    """
    return reduce(lambda x, y: x - y, lista_numeros)

# Ejemplo de uso
numeros = [4, 6, 200, 4, 605]
resultado = diferencia_total(numeros)
print("La diferencia total de los valores es:", resultado)
```

```python
"""
Ejercicico 25. Crea una función que cuente el número de caracteres en una cadena de texto dada.
"""
def contar_caracteres(cadena):
    """
    Cuenta el número total de caracteres en una cadena de texto.

    Args:
        cadena (str): Texto introducido por el usuario.

    Returns:
        int: Número total de caracteres en la cadena.
    """
    return len(cadena)

# Ejemplo de uso
texto = input("Introduce una cadena de texto: ")
resultado = contar_caracteres(texto)
print("El número total de caracteres es:", resultado)
```

```python
"""
Ejercicio 26. Crea una función lambda que calcule el resto de la división entre dos números dados.
"""
# Función lambda que calcula el resto (módulo) de una división
resto = lambda x, y: x % y

try:
    num1 = int(input("Introduce el primer número: "))
    num2 = int(input("Introduce el segundo número: "))

    # Verificamos si el divisor es cero
    if num2 == 0:
        raise ZeroDivisionError("No se puede dividir entre cero.")

    resultado = resto(num1, num2)
    print(f"El resto de dividir {num1} entre {num2} es: {resultado}")

except ValueError:
    print("Error: Debes introducir solo números enteros.")
except ZeroDivisionError as e:
    print(f"Error de división: {e}")
except Exception as e:
    print(f"Ocurrió un error inesperado: {e}")
```

```python
"""
Ejercicio 27. Crea una función que calcule el promedio de una lista de números.
"""
def calcular_promedio(numeros):
    """
    Calcula el promedio de una lista de números.

    Args:
        numeros (list): Lista de números.

    Returns:
        float: Promedio de los números en la lista.
    """
    if not numeros:
        raise ValueError("La lista está vacía. No se puede calcular el promedio.")
    
    return sum(numeros) / len(numeros)

# Ejemplo de uso con manejo de errores
try:
    entrada = input("Introduce una lista de números separados por espacios: ")
    lista = list(map(float, entrada.split()))
    
    promedio = calcular_promedio(lista)
    print("El promedio es:", promedio)

except ValueError as e:
    print(f"Error de valor: {e}")
except Exception as e:
    print(f"Ocurrió un error inesperado: {e}")
```

```python
"""
Ejercicio 28. Crea una función que busque y devuelva el primer elemento duplicado en una lista dada.
"""
def primer_duplicado(lista):
    """
    Busca y devuelve el primer elemento duplicado en una lista.

    Args:
        lista (list): Lista de elementos.

    Returns:
        any: Primer elemento duplicado encontrado. Si no hay duplicados, devuelve None.
    """
    vistos = set()
    for elemento in lista:
        if elemento in vistos:
            return elemento
        vistos.add(elemento)
    return None

# Ejemplo de uso
lista_ejemplo = [2, 5, 3, 4, 5, 3]
resultado = primer_duplicado(lista_ejemplo)

if resultado is not None:
    print(f"El primer elemento duplicado es: {resultado}")
else:
    print("No se encontraron elementos duplicados.")
```

```python
"""
Ejercicio 29. Crea una función que convierta una variable en una cadena de texto y enmascare todos los caracteres con el
carácter '#', excepto los últimos cuatro.
"""
def enmascarar_variable(valor):
    """
    Convierte un valor en cadena de texto y enmascara todos los caracteres con '#' 
    excepto los últimos cuatro.

    Args:
        valor (any): Valor a convertir y enmascarar.

    Returns:
        str: Cadena enmascarada.
    """
    cadena = str(valor)  # Convertimos el valor a cadena
    longitud = len(cadena)
    
    if longitud <= 4:
        # Si tiene 4 o menos caracteres, no enmascaramos nada
        return cadena
    
    # Enmascaramos todos los caracteres excepto los últimos 4
    enmascarado = '#' * (longitud - 4) + cadena[-4:]
    return enmascarado

# Ejemplo de uso
variable = input("Introduce un valor a enmascarar: ")
resultado = enmascarar_variable(variable)
print("Valor enmascarado:", resultado)
```

```python
"""
Ejercicio 30. Crea una función que determine si dos palabras son anagramas, es decir, si están formadas por las mismas letras
pero en diferente orden.
"""
def son_anagramas(palabra1, palabra2):
    """
    Determina si dos palabras son anagramas.

    Args:
        palabra1 (str): Primera palabra.
        palabra2 (str): Segunda palabra.

    Returns:
        bool: True si son anagramas, False en caso contrario.
    """
    # Convertimos a minúsculas y eliminamos espacios para comparación justa
    p1 = palabra1.replace(" ", "").lower()
    p2 = palabra2.replace(" ", "").lower()

    # Comparamos las palabras ordenando sus letras
    return sorted(p1) == sorted(p2)

# Ejemplo de uso
palabra_a = input("Introduce la primera palabra: ")
palabra_b = input("Introduce la segunda palabra: ")

if son_anagramas(palabra_a, palabra_b):
    print(f'"{palabra_a}" y "{palabra_b}" son anagramas.')
else:
    print(f'"{palabra_a}" y "{palabra_b}" NO son anagramas.')
```

```python
"""
Ejercicio 31. Crea una función que solicite al usuario ingresar una lista de nombres y luego solicite un nombre para buscar en
esa lista. Si el nombre está en la lista, se imprime un mensaje indicando que fue encontrado, de lo contrario, se
lanza una excepción.
"""
# Definimos una excepción personalizada
class NombreNoEncontradoError(Exception):
    """Excepción que se lanza cuando un nombre no se encuentra en la lista."""
    pass

def buscar_nombre():
    """
    Solicita una lista de nombres y un nombre para buscar. 
    Imprime un mensaje si se encuentra, o lanza una excepción si no.
    """
    # Solicitar lista de nombres
    nombres = input("Introduce una lista de nombres separados por comas: ").split(",")
    # Limpiamos espacios en blanco
    nombres = [nombre.strip() for nombre in nombres]

    # Solicitar nombre a buscar
    nombre_a_buscar = input("Introduce el nombre a buscar: ").strip()

    # Verificar si el nombre está en la lista
    if nombre_a_buscar in nombres:
        print(f'El nombre "{nombre_a_buscar}" fue encontrado en la lista.')
    else:
        raise NombreNoEncontradoError(f'El nombre "{nombre_a_buscar}" NO se encuentra en la lista.')

# Ejemplo de uso con manejo de excepción
try:
    buscar_nombre()
except NombreNoEncontradoError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"Ocurrió un error inesperado: {e}")
```

```python
"""
Ejercicio 32. Crea una función que tome un nombre completo y una lista de empleados, busque el nombre completo en la lista y
devuelve el puesto del empleado si está en la lista, de lo contrario, devuelve un mensaje indicando que la persona
no trabaja aquí.
"""
def buscar_puesto(nombre_completo, empleados):
    """
    Busca el puesto de un empleado en una lista de empleados.

    Args:
        nombre_completo (str): Nombre completo del empleado a buscar.
        empleados (list of dict): Lista de empleados con 'nombre' y 'puesto'.

    Returns:
        str: Puesto del empleado si se encuentra, o mensaje indicando que no trabaja aquí.
    """
    for empleado in empleados:
        if empleado['nombre'].lower() == nombre_completo.lower():
            return f"{nombre_completo} ocupa el puesto de {empleado['puesto']}."
    return f"{nombre_completo} no trabaja aquí."

# Ejemplo de uso
empleados = [
    {'nombre': 'Ana Perez', 'puesto': 'Gerente'},
    {'nombre': 'Luis Gomez', 'puesto': 'Desarrollador'},
    {'nombre': 'Marta Lopez', 'puesto': 'Diseñadora'}
]

nombre_buscar = input("Introduce el nombre completo del empleado: ")
resultado = buscar_puesto(nombre_buscar, empleados)
print(resultado)
```

```python
"""
Ejercicio 33. Crea una función lambda que sume elementos correspondientes de dos listas dadas.
"""
# Función lambda que suma elementos correspondientes de dos listas
sumar_listas = lambda lista1, lista2: list(map(lambda x, y: x + y, lista1, lista2))

# Ejemplo de uso
lista_a = [1, 2, 3]
lista_b = [4, 5, 6]

resultado = sumar_listas(lista_a, lista_b)
print("Suma de elementos correspondientes:", resultado)
```
